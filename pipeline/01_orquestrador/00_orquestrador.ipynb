```python
# ==============================
# Notebook Orquestrador - Manual + Job
# ==============================

# -----------------------------
# 0 Teste manual: define input_path
# -----------------------------
input_path = "/Volumes/saas_project/core/landing/crypto_sentiment_prediction_dataset.csv"

# -----------------------------
# 1 Recebe input_path do Job (caso esteja rodando via Job)
# -----------------------------
job_input_path = dbutils.widgets.get("input_path")

# -----------------------------
# 2 Decide qual input_path usar
# -----------------------------
if job_input_path:
    input_path_final = job_input_path
    print(f" Usando input_path do Job: {input_path_final}")
elif input_path:
    input_path_final = input_path
    print(f" Usando input_path definido manualmente: {input_path_final}")
else:
    raise ValueError(" input_path não fornecido nem definido manualmente!")

# -----------------------------
# 3 Chama ingestão Bronze passando input_path_final
# -----------------------------
dbutils.notebook.run(
    "./01_ingestao_bronze",
    0,
    {"input_path": input_path_final}  # chave = nome do widget no Bronze
)

# -----------------------------
# 4 Chama os demais notebooks do pipeline
# -----------------------------
dbutils.notebook.run("./02_bronze_para_silver", 0)
dbutils.notebook.run("./03_SilverRobusto", 0)
dbutils.notebook.run("./04_silver_para_gold", 0)
dbutils.notebook.run("./05_qualidade_e_alertas", 0)



# -----------------------------
# 5 Mensagem final
# -----------------------------
print(" Pipeline concluído! Manual ou Job ")
```
